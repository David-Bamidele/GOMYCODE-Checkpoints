```python
import numpy as np
import pandas as pd 
```

```python
df = pd.read_csv('Loan_prediction_dataset.csv')
df
```

### Find the **mean**, **median**, and **standard deviation** of the loan amounts

#### mean

```python
loan_amounts = df['LoanAmount'].to_numpy()
```

```python
mean_loan_amount = np.nanmean(loan_amounts)
print(f"Mean Loan Amount: {mean_loan_amount}")
```

#### median

```python
median_loan_amount = np.nanmedian(loan_amounts)
print(f"Median Loan Amount: {median_loan_amount}")
```

#### standard deviation

```python
std_dev_loan_amount = np.nanstd(loan_amounts)
print(f"Standard Deviation of Loan Amount: {std_dev_loan_amount}")
```
